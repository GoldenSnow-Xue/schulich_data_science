```python
# Divide results into three groups by injury level: Green, Yellow and Red;. 

# Create a new variable called classification based on 3 conditions: 

# Let’s make death > 0 with no matter how many injuries. Tag as ‘Red’
# death = 0. With injuries. Tag as ‘Yellow’
# Otherwise. Tag as ‘Green’. Code are listed below:
```

```python
conditions = [(df3['DEATH'] > 0), (df3['DEATH'] == 0) & (df3['INJURIES'] > 0), 
              (df3["DEATH"] < 1) & (df3["INJURIES"] < 1)]
choices = ["Red","Yellow","Green"] 
```

```python
df3['Classification'] = np.select(conditions, choices)
```

```python
df3['Classification'].value_counts()
```
