# 2. Hypothesis Testing

### 2.1 Seasonal Price Variation

Hypothesis: Prices vary significantly across different seasons.

Testing Approach: Group the data by seasons or months and use ANOVA to test for differences in average prices across these time periods.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

seasonal_price_means = df.groupby('Season')['price'].mean()

# ANOVA Test
model = ols('price ~ C(Season)', data=df).fit()
anova_results = sm.stats.anova_lm(model, typ=2)

print(anova_results)
```

The very low p-value (0.0) strongly suggests that there are statistically significant differences in mean prices among the different seasons.
The high F-statistic reinforces that the seasonal effect on prices is not only statistically significant but also practically meaningful.

**Implications for Time-Based Dynamic Pricing:**

Seasonal Pricing Strategy: The results support a dynamic pricing strategy that varies prices by season. You can adjust prices based on the season to optimize revenue, taking advantage of higher demand or price sensitivity in certain seasons.

### 2.2 Price Variation by City

Hypothesis: Prices differ significantly between cities.

Testing Approach: Use ANOVA to compare the average prices across different cities.

```python
model = ols('price ~ C(city)', data=df).fit()
anova_results = sm.stats.anova_lm(model, typ=2)

print(anova_results)
```

The extremely low p-value and high F-statistic strongly indicate that there are significant differences in mean prices across different cities. This suggests that the city is a significant factor influencing price variation in your dataset.

**Implications for Location-Based Dynamic Pricing:**

For a location-based dynamic pricing strategy, these results suggest that adjusting prices according to the city can be a valid approach. It confirms that different cities may have different pricing dynamics possibly due to factors like local demand, cost of living, or competition.

### 2.4 Price Differences During Promotions

Hypothesis: The prices of products are significantly different when they are on promotion compared to when they are not.

Testing Approach: Use a t-test (if the data is normally distributed) or a non-parametric test like the Mann-Whitney U test (for non-normal distributions) to compare prices when onPromo is true versus when it is false.

```python
from scipy import stats

promo_prices = df[df['onPromo'] == True]['price']
non_promo_prices = df[df['onPromo'] == False]['price']

# Normality Test
print("Promo Prices Normality Test:", stats.shapiro(promo_prices))
print("Non-Promo Prices Normality Test:", stats.shapiro(non_promo_prices))

# Independent T-Test
t_stat, p_value = stats.ttest_ind(promo_prices, non_promo_prices, equal_var=False)
print("T-Test Result:", t_stat, "P-Value:", p_value)
```

The warning from the Shapiro-Wilk test (p-value may not be accurate for N > 5000) suggests that the normality test may not be reliable due to a large sample size. However, the very low p-values (0.0) in both the promotional and non-promotional groups suggest that the price distributions do not follow a normal distribution.

```python
u_stat, p_value = stats.mannwhitneyu(promo_prices, non_promo_prices, alternative='two-sided')
print("Mann-Whitney U Test Result:", u_stat, "P-Value:", p_value)
```

Mann-Whitney U Statistic: The value of 24253327869.0 itself doesn't provide much interpretive value without context, but it's used internally in the calculation of the p-value.

P-Value: A p-value of 0.0 (or very close to 0) means that the probability of observing such a large difference in prices between the two groups, purely by chance, is extremely low. This allows you to confidently reject the null hypothesis that there is no difference in pricing between products on promotion and those not on promotion.

**Implications for Dynamic Pricing:**

Promotional Pricing Impact: This result supports the idea that promotional periods have a distinct impact on pricing strategies. We can leverage this insight by tailoring prices more aggressively during promotions to maximize sales and revenue.

Data-Driven Strategy: The clear difference in pricing strategies between promotional and non-promotional periods can be incorporated into the dynamic pricing model, using promotions as a key factor for price adjustments.
