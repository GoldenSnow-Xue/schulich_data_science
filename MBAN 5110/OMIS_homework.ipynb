##### 2. Run the OLS model

```python
# Define the Dependent Variable and the Explanatory Variables
dependent_var = df['Dependent Var']
explanatory_vars = df.drop('Dependent Var', axis=1)
```

```python
# Run the OLS Regression
OLS_Model = sm.OLS(dependent_var, explanatory_vars).fit()
OLS_Model.summary()
```

```python
# Add a constant to the model (intercept)
explanatory_vars_updated = sm.add_constant(explanatory_vars)
OLS_Model_updated = sm.OLS(dependent_var, explanatory_vars_updated).fit()
OLS_Model_updated.summary()
```

##### **Part B**

##### 1. Use error values from the OLS model to calculate their standard deviation and autocorrelation values for the first three lags. 

```python
# Extract the residuals (error values) from the OLS Model
residuals = OLS_Model_updated.resid
```

```python
# Calculate the standard deviation of residuals
std_dev = residuals.std()
print(f'Standard Deviation of Residuals: {std_dev}')
```

```python
# Calculate the autocorrelation values for the first three lags
lag_1 = residuals.autocorr(lag=1)
lag_2 = residuals.autocorr(lag=2)
lag_3 = residuals.autocorr(lag=3)
```

##### 2. Run the GLS Model

```python
# Import Relevant Libraries
from scipy.linalg import toeplitz
```

```python
# Create the Toeplitz matrix
toeplitz_values = [1, lag_1, lag_2, lag_3]
toeplitz_matrix = toeplitz(toeplitz_values)
```

```python
# If your data has more than 4 observations, need to extend the Toeplitz matrix to match the number of observations
nobs = len(residuals)
toeplitz_matrix = toeplitz(np.r_[1, lag_1, lag_2, lag_3, np.zeros(nobs-4)])
```

```python
# Run the GLS Model
GLS_Model = sm.GLS(dependent_var, explanatory_vars_updated, sigma=toeplitz_matrix).fit()
GLS_Model.summary()
```

##### **Part C**

##### 1. Split the dataset into two as the training and test sets (test size = 0.5)

```python
# Import Relevant Libraries
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(explanatory_vars, dependent_var, test_size=0.5, random_state=42)
```

##### 2. Run the Lasso model with alpha=1 and estimate the coefficients using the training set. 

```python
# Import Relevant Libraries
from sklearn.linear_model import Lasso
```

```python
# Run the Lasso Model with alpha=1
Lasso_Model = Lasso(alpha=1).fit(X_train, y_train)
```

```python
# Estimate the coefficients using the training set
Lasso_Model.coef_
```

##### 3. Calculate the mean absolute percentage error using the test set

```python
# Import Relevant Libraries
from sklearn.metrics import mean_absolute_error
```

```python
# Make predictions using the test set
y_pred = Lasso_Model.predict(X_test)
```

```python
# Calculate the Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print('Mean Absolute Percentage Error (MAPE):', mape)
```

##### 4. Find an approximate value for alpha that minimizes the mean absolute percentage error.

```python
# Function to calculate MAPE
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    non_zero_elements = y_true != 0
    return np.mean(np.abs((y_true[non_zero_elements] - y_pred[non_zero_elements]) / y_true[non_zero_elements])) * 100
```

```python
# List of alphas to try
alphas = np.logspace(-6, 0, 100)
```

```python
# Placeholder for the best MAPE and corresponding alpha
best_mape = float('inf')
best_alpha = None
```

```python
# Loop over alphas
for alpha in alphas:
    # Create Lasso model
    lasso_model = Lasso(alpha=alpha, max_iter=10000)
    
    # Fit Lasso model on the training data
    lasso_model.fit(X_train, y_train)
    
    # Make predictions on the test data
    y_pred = lasso_model.predict(X_test)
    
    # Calculate MAPE
    mape = mean_absolute_percentage_error(y_test, y_pred)
    
    # Update the best MAPE and alpha if necessary
    if mape < best_mape:
        best_mape = mape
        best_alpha = alpha

print('Best Alpha:', best_alpha)
print('Best MAPE:', best_mape)
```

```python
# Create dataframe from the provided data
df = pd.DataFrame({
    'Month': list(range(1, 26)),
    'Demand': [100, 112, 107, 103, 91, 85, 84, 85, 79, 81, 134, 86, 99, 89, 111, 114, 118, 163, 193, 143, 144, 202, 158, 160, 144],  # fill in the demand values here
    'Advance_demand': [71, 30, 75, 64, 41, 51, 42, 51, 57, 49, 134, 52, 99, 56, 81, 79, 73, 163, 193, 99, 91, 202, 105, 101, 96]  # fill in the advance demand values here
})
```

```python
# Create lagged demand features
for i in range(1, 4):  # for demonstration purposes, we're considering 3 lags
    df[f'Demand_lag_{i}'] = df['Demand'].shift(i)
```

```python
# Create difference between Demand and Advance_demand
df['Demand_diff'] = df['Demand'] - df['Advance_demand']
```

```python
# Split the data
train = df[df['Month'] <= 20]
test = df[df['Month'] > 20]

X_train = train.drop(columns=['Month', 'Demand'])
y_train = train['Demand']

X_test = test.drop(columns=['Month', 'Demand'])
y_test = test['Demand']
```

```python
# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Impute the missing values
imputer = SimpleImputer(strategy="mean")
X_train_scaled = imputer.fit_transform(X_train_scaled)
X_test_scaled = imputer.transform(X_test_scaled)
```

```python
# Train Lasso model
lasso = Lasso(alpha=1.0)
lasso.fit(X_train_scaled, y_train)
```

```python
# Predict on test set
y_pred = lasso.predict(X_test_scaled)
```

```python
# Calculate error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```
