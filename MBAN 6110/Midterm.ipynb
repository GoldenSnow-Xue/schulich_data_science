```python
units_by_region = df.groupby('location')['checkout_status'].aggregate(['sum','mean','median','count'])
units_by_region
```

```python
df['success'] = None
df['fail'] = None
```

```python
df.loc[df['checkout_status'] == 1, 'success'] = df['checkout_status']
df.loc[df['checkout_status'] == 0, 'fail'] = df['checkout_status']
```

```python
df.info()
```

```python
by_location = df.groupby('location')['success'].aggregate(['count'])
by_location
```

```python
by_device = df.groupby('device')['success'].aggregate(['count'])
by_device
```

```python
by_age = df.groupby('age')['success'].aggregate(['count'])
by_age
```

```python
by_time_spent = df.groupby('time_spent')['success'].aggregate(['count'])
by_time_spent
```
