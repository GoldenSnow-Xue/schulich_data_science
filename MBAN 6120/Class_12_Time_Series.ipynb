```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = yf.download('META', start='2017-01-01', end='2023-12-07')
data
```

```python
data = data['Adj Close']
data
```

```python
data.plot()
```

```python
data.describe()
```

```python
plt.hist(data)
```

bimodal - two peaks

```python
data = data.to_frame()
data
```

```python
data['7_daysMA'] = data['Adj Close'].rolling(window=7).mean()
data
```

```python
data['7_daysMA'].plot()
```

```python
data['30_daysMA'] = data['Adj Close'].rolling(window=30).mean()
data['30_daysMA'].plot()
```

```python
data.reset_index(inplace=True)
data
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data
```

```python
data.set_index('Date', inplace=True)
data
```

```python
seasonal_df = data.pivot_table(values='Adj Close', aggfunc='mean', index=data.index.month, columns=data.index.year)
seasonal_df
```

```python
seasonal_df.plot()
```

```python
pd.plotting.lag_plot(data['Adj Close'], lag=7)
```

```python
data['7_day_lag'] = data['Adj Close'].shift(7)
data
```

```python
plt.scatter(data['7_day_lag'], data['Adj Close'])
```

Business days (not including weekends)

Larger lag, less correlation. (BUT also depends on the industry selected, more stable the industry, higher the correlation)
